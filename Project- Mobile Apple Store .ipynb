```python
import matplotlib.pyplot as plt

avg_StudyTimeWeekly = df.groupby('price')['id'].mean()
plt.bar(avg_StudyTimeWeekly.index, avg_StudyTimeWeekly)  # Corrected this line
plt.xlabel('price')
plt.ylabel('Average id')
plt.title('Average id by price (Matplotlib)')
plt.xticks(rotation=45)
plt.show()
```
